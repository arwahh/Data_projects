```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier


X = np.array([1,2,3,6,7,8]).reshape(-1,1)
y = np.array([0,0,0,1,1,1])


model = KNeighborsClassifier(n_neighbors=3)
model.fit(X, y)


print("Predicted Class:", model.predict([[5]]))
```

```python

new_points = np.array([2, 4, 7]).reshape(-1,1)
print(model.predict(new_points))
```

```python
for k in [1,3,5]:
    model = KNeighborsClassifier(n_neighbors=k)
    model.fit(X, y)
    print(f"K={k}, Prediction:", model.predict([[5]]))
```

```python
from sklearn.neighbors import KNeighborsRegressor

X = np.array([1,2,3,6,7]).reshape(-1,1)
y = np.array([10,12,14,20,22])

model = KNeighborsRegressor(n_neighbors=2)
model.fit(X, y)

print("Predicted Value:", model.predict([[5]]))
```

```python
from sklearn.neighbors import KNeighborsClassifier
X=[
    [180,7],
    [200,7.5],
    [250,8],
    [300,8.5],
    [330,9],
    [360,9.5]

]
# 0=Apple, 1 =Orange
y=[0,0,0,1,1,1]
model = KNeighborsClassifier(n_neighbors=3)
model.fit(X,y)
weight= float(input("enter the weight in grams:"))
size=float(input("enter the size in cm:"))
prediction=model.predict([[weight,size]])[0]
if prediction==0:
  print("This is likely an apple")
else:
  print("This is likely an orange")
```
